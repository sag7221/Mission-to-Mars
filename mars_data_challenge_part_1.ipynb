```python
# Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager

executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the Mars NASA news site
url = 'https://redplanetscience.com'
browser.visit(url)
```

```python
html = browser.html
html_soup = soup(html, 'html.parser')
print(html_soup)
```

```python
div_list = html_soup.find_all('div', class_='list_text' )
print(div_list[0])
dict_elem = {}

dict_elem['title'] = div_list[0].select_one('div.content_title').text
dict_elem['preview'] = div_list[0].select_one('div.article_teaser_body').text
print(dict_elem)
```

```python
elem_list = []
for div in div_list:
    dict_elem = {}
    dict_elem['title'] = div.select_one('div.content_title').text
    dict_elem['preview'] = div.select_one('div.article_teaser_body').text
    elem_list.append(dict_elem)
print(elem_list)
    
```

```python
import json
json_string = json.dumps(elem_list)
with open('mars_news.json', 'w') as f:
    json.dump(json_string, f)

#     End of DELIVERABLE 1
```
